```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#importing dataset
store=pd.read_csv('SampleSuperstore.csv')
```

```python
discount =store['Discount'].sum()
discount
```

```python
#getting states and counts of each state
store['State'].value_counts()
```

```python
#getting top 10 sales
Top_10= store.groupby("State").Sales.sum().nlargest(n=10)
Top_10
```

```python
#plotting top 10 sales in each state
plt.figure(figsize = (10, 7))
plt.style.use('grayscale')
Top_10.plot(kind ='bar',color="pink",figsize =(8,6), fontsize =14)
plt.xlabel("States", fontsize =14)
plt.ylabel("Total Sales",fontsize =14)
plt.title('Top 10 Sales per State',fontsize =15)
plt.show()
```

```python
store.corr()
```

```python
#plotting heatmap 
plt.figure(figsize = (10, 7))
sns.heatmap(store.corr(),annot=True)
```

```python
#getting counts for each category
df2=store['Category'].value_counts()
df2
```

```python
#plotting different categories w.r.t count
plt.figure(figsize = (10, 7))
sns.countplot(x='Category',data=store,palette='tab20')
```

```python
#plotting all the subcategories
plt.figure(figsize = (14, 6))
sns.countplot(x='Sub-Category',data=store,palette='tab20')
```

```python
# We will plot barplot of Category in terms of Profit
plt.figure(figsize = (10, 7))
sns.barplot(x='Category',y='Profit',data=store,palette='tab20')
```

```python
store.groupby(['Category'])['Discount'].sum()
```

```python
# We will plot barplot of Category in terms of Discount
plt.figure(figsize = (10, 7))
sns.barplot(x='Category',y='Discount',data=store,palette='tab20')
```

```python
store.groupby(['Category'])['Sales'].sum()
```

```python
## We will plot barplot of Category in terms of Sales
plt.figure(figsize = (10, 7))
sns.barplot(x='Category',y='Sales',data=store,palette='tab20')
```

```python
# We will plot Pairplot on the basis of Category
sns.pairplot(store,hue='Category',palette='tab20')
```

```python
store.groupby("State").Profit.agg(["sum","mean",'median','std','count','var']).head()
```

```python
# We will plot countplot of Sub-Categoreis as per the Region
plt.figure(figsize = (14, 6))
sns.countplot(x = 'Sub-Category', hue = 'Region', data = store, palette ='tab10')
```

```python
# We will plot countplot of Categoreis as per the Region
plt.figure(figsize = (10, 6))
sns.countplot(x = 'Category', hue = 'Region', data = store, palette = 'tab10')
plt.xticks(rotation = 90, fontsize = 10)
plt.yticks(fontsize = 10)
plt.xlabel('Categories',fontsize = 20)
plt.legend(loc = 5, fontsize = 10)
plt.show()
```

```python
## We will plot countplot of Segments 
plt.figure(figsize = (10, 8))
sns.countplot(x = store["Segment"],palette='tab20')
```

```python
## We will plot countplot of States
plt.figure(figsize=(14,4))
sns.countplot(x=store['State'])
plt.xticks(rotation=90,fontsize = 14)
plt.yticks(fontsize=14)
plt.show()
```

```python
#We will plot barplot for Sub-Category and Sales
plt.figure(figsize=[14,8])
ax = sns.barplot(x="Sub-Category", y="Sales", data=store, palette="tab10",edgecolor='black')
plt.xlabel("Subcategories",fontsize=15)
plt.ylabel("Sales",fontsize=15)
plt.xticks(rotation=90,fontsize = 14)
```

```python
#We will plot barplot for Sub-Category and Profit
plt.figure(figsize=[14,8])
ax = sns.barplot(x="Sub-Category", y="Profit", data=store, palette="tab10",edgecolor='black')
plt.xlabel("Subcategory", fontsize=15)
plt.ylabel("Profit",fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

```python
#We will plot barplot for Sub-Category and discount
plt.figure(figsize=[14,8])
ax = sns.barplot(x="Sub-Category", y="Discount", data=store, palette="tab10",edgecolor='black')
plt.xlabel("Subcategory", fontsize=15)
plt.ylabel("dis",fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

```python
#plotting a barplot for profits per state
plt.figure(figsize=[14,8])
ax = sns.barplot(x="State", y="Profit", data=store, palette="tab20",edgecolor='black')
plt.xticks(rotation=90, fontsize=16)
plt.yticks(fontsize=15)
plt.title("Profit per State",fontsize=24)
plt.xlabel("States",fontsize=20)
plt.ylabel("Profit",fontsize=20)
plt.tight_layout()
```

```python
##plotting a barplot for discounts per state
plt.figure(figsize=[14,8])
ax = sns.barplot(x="State", y="Discount", data=store, palette="tab20",edgecolor='black')
plt.xticks(rotation=90, fontsize=16)
plt.yticks(fontsize=15)
plt.title("discount per State",fontsize=24)
plt.xlabel("States",fontsize=20)
plt.ylabel("discount",fontsize=20)
plt.tight_layout()
```

```python
##plotting a barplot for sales per state
plt.figure(figsize=[14,8])
ax = sns.barplot(x="State", y="Sales", data=store, palette="tab20",edgecolor='black')
plt.xticks(rotation=90, fontsize=16)
plt.yticks(fontsize=15)
plt.title("sales per State",fontsize=24)
plt.xlabel("States",fontsize=20)
plt.ylabel("sales",fontsize=20)
plt.tight_layout()
```
